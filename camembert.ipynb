# Création de diagrammes circulaires (camemberts)

```python
%matplotlib inline
import matplotlib.pyplot as plt


serie_1 = [93, 6, 0, 0, 0, 0, 0]
serie_2 = [0, 0, 32, 30, 15, 14, 9]
etiquettes = ['Hydrogène', 'Hélium', 'Fer', 'Oxygène', 'Silicium', 'Magnésium', 'Autres']
etiquettes_courtes = ['H', 'He', 'Fe', 'O', 'Si', 'Mg', 'Autres']
titre_1 = "Soleil"
titre_2 = "Terre"


plt.rcParams['figure.figsize'] = (9, 5)

plt.subplot (1,2,1)
plt.pie(
        serie_1, 
        labels = [etiquettes_courtes[t] if serie_1[t]>0 else '' for t in range(len(etiquettes_courtes))], 
        labeldistance = 1.15, 
        autopct=lambda p: '{:.1f}%'.format(round(p)) if p > 0 else '', 
        shadow = True
        )
plt.title(titre_1)

plt.subplot (1,2,2)
plt.pie(
        serie_2, 
        labels = [etiquettes_courtes[t] if serie_2[t]>0 else '' for t in range(len(etiquettes_courtes))], 
        labeldistance = 1.15, 
        autopct=lambda p: '{:.1f}%'.format(round(p)) if p > 0 else '', 
        shadow = True
        )
plt.title(titre_2)

plt.legend(labels = etiquettes, bbox_to_anchor=(1.0, 1.0), loc='upper left')
plt.tight_layout()


plt.savefig("camembert.png")
```
