```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
fig = plt.figure()

def anime(i):
    
    x=i%100
    y=0.5
    if x==0:
        fig.clear()
    plt.plot([x],[y],marker='+',color="orange")
    plt.xlim((0,100))
    plt.ylim((0,1))

simulation=animation.FuncAnimation(fig,anime,interval=20,blit=True,) #appelle la fonction anime toutes les 50 ms avec un indice i de plus en plus grand

plt.show()
```
