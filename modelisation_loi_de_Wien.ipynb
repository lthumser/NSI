```python
%matplotlib notebook
import matplotlib.pyplot as plt 
```

## loi de Wien : valeurs expérimentales

```python
T = [3500,4000,4500,5000,6000]   # température en K
un_sur_T = [1/i for i in T]
Lambda = [850,750,630,580,490] # longueur d'onde en nm

x,y = un_sur_T,Lambda
```

## Modélisation

```python
from scipy.optimize import curve_fit

def lineaire(x,a):
    return a*x

params,covar = curve_fit(lineaire,x,y)

m=[]

for val in x:
    m.append(lineaire(val,params))

print("Constante = ",*params*1E-9,"m·K")
```

```python
titre_abscisse = "1/T ($K^{-1}$)"
titre_ordonnee = "$λ_{max}$ (nm)"

plt.plot(x,y,"+",label="valeurs expérimentales")
plt.plot(x,m,label="modele {} = {:.3f} × {}".format(titre_abscisse, *params,titre_ordonnee))


plt.legend()
plt.title("modélisation de la loi de Wien")
plt.xlabel(titre_abscisse)
plt.ylabel(titre_ordonnee)
plt.show()
```
